# The first step is importing the necessary libraries, such as BeautifulSoup, Requests and CSV

```python
from bs4 import BeautifulSoup as bs
import requests
import csv 
```

## Sending a GET request to the given url and extracting the html content of the response 

```python
url = 'https://www.learnit3d.fi/list-of-european-countries-and-capitals/'
```

```python
response = requests.get(url)
response
```

```python
html = response.content
print(html)
```

```python
soup = bs(html, 'lxml')
print(soup)
```

```python
soup.tbody
```

## Find all the rows within the tbody
## Iterate over each row and extract the 'column-1' and 'column-2' texts

```python
rows = soup.tbody.find_all('tr')


for row in rows:
    column_1_text = row.find('td', class_='column-1').text
    column_2_text = row.find('td', class_='column-2').text
    print(column_1_text, column_2_text)
```

## Specify the file path where you want to save the CSV file, open the CSV file in write mode.
## Then create a CSV writer object, a header row, and finally iterate over each row and extract the 'column-1' and 'column-2' texts.

```python

csv_file_path = '/home/dataguy/hobby_project/city_and_country.csv'


with open('/home/dataguy/hobby_project/city_and_country.csv', 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['Country', 'City'])  # Modify the headers if needed

    
    for row in rows:
        column_1_text = row.find('td', class_='column-1').text
        column_2_text = row.find('td', class_='column-2').text
        csv_writer.writerow([column_1_text, column_2_text])
```
